```python
# ─── Imports & Data Loading ──────────────────────────────
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from kmeans_alternate import KMeansAlternate
from kmeans_custom import KMeansCustom
import time

%matplotlib inline
plt.rcParams['figure.dpi'] = 120

# ─── Load & Preprocess ──────────────────────────────────
iris = pd.read_csv('data/iris.csv')
ai   = pd.read_csv('data/ai_index.csv')
eq   = pd.read_csv('data/earthquakes.csv')

def preprocess_iris(df):
    df = df.drop(columns=['species']).dropna()
    return StandardScaler().fit_transform(df)

def preprocess_ai(df):
    df = df.select_dtypes(include=[np.number]).dropna()
    return StandardScaler().fit_transform(df)

def preprocess_eq(df):
    df = df[['magnitude','depth','latitude','longitude']].dropna()
    return StandardScaler().fit_transform(df)

iris_data = preprocess_iris(iris)
ai_data   = preprocess_ai(ai)
eq_data   = preprocess_eq(eq)

datasets = {
    'Iris':        (iris_data, 3),
    'AI Index':    (ai_data, 3),
    'Earthquakes': (eq_data, 3),
}

print("Datasets loaded:")
for name, (d, k) in datasets.items():
    print(f"  {name}: {d.shape[0]} samples, {d.shape[1]} features, k={k}")
```

```python
# ══════════════════════════════════════════════════════════
# PART 1 — CUSTOM ALGORITHM STEP-BY-STEP
# ══════════════════════════════════════════════════════════
# Our custom KMeans has two main phases:
#   Phase A: ESPI Initialization (smart centroid seeding)
#   Phase B: Drift-based Iterative Refinement
#
# This cell visualizes Phase A step by step for each dataset.

def visualize_espi_init(data, dataset_name, k=3):
    """Show the 4 steps of ESPI initialization."""
    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)
    colors = plt.cm.Set1(np.linspace(0, 0.9, k))

    fig, axes = plt.subplots(1, 4, figsize=(22, 5))
    fig.suptitle(f'Phase A: ESPI Initialization — {dataset_name}  (k={k})',
                 fontsize=15, fontweight='bold', y=1.02)

    # ── Step 1: Raw Data ─────────────────────────────────
    axes[0].scatter(data_2d[:, 0], data_2d[:, 1],
                    c='steelblue', alpha=0.5, s=30, edgecolors='white',
                    linewidth=0.3)
    axes[0].set_title('Step 1: Raw Data\n(PCA projected to 2D)', fontsize=11)
    axes[0].set_xlabel('PC1'); axes[0].set_ylabel('PC2')

    # ── Step 2: Find dominant dimension ──────────────────
    variances = np.var(data, axis=0)
    dominant_dim = np.argmax(variances)
    dom_vals = data[:, dominant_dim]
    sc = axes[1].scatter(data_2d[:, 0], data_2d[:, 1],
                         c=dom_vals, cmap='coolwarm', alpha=0.7, s=30,
                         edgecolors='white', linewidth=0.3)
    plt.colorbar(sc, ax=axes[1], label=f'Feature {dominant_dim} value')
    axes[1].set_title(f'Step 2: Find Dominant Dimension\n'
                      f'dim={dominant_dim}, variance={variances[dominant_dim]:.2f}',
                      fontsize=11)
    axes[1].set_xlabel('PC1'); axes[1].set_ylabel('PC2')

    # ── Step 3: Sort & Partition ─────────────────────────
    sorted_indices  = np.argsort(data[:, dominant_dim])
    sorted_data     = data[sorted_indices]
    sorted_data_2d  = data_2d[sorted_indices]
    partitions_2d   = np.array_split(sorted_data_2d, k)
    partitions_orig = np.array_split(sorted_data, k)

    for i, part in enumerate(partitions_2d):
        axes[2].scatter(part[:, 0], part[:, 1],
                        c=[colors[i]], alpha=0.6, s=30,
                        edgecolors='white', linewidth=0.3,
                        label=f'Partition {i+1} ({len(part)} pts)')
    axes[2].set_title(f'Step 3: Sort by dim {dominant_dim}\n'
                      f'then split into {k} equal groups',
                      fontsize=11)
    axes[2].set_xlabel('PC1'); axes[2].set_ylabel('PC2')
    axes[2].legend(fontsize=8)

    # ── Step 4: Centroids = mean of each partition ───────
    centroids_orig = np.array([p.mean(axis=0) for p in partitions_orig])
    centroids_2d   = pca.transform(centroids_orig)

    for i, part in enumerate(partitions_2d):
        axes[3].scatter(part[:, 0], part[:, 1],
                        c=[colors[i]], alpha=0.25, s=25)
    for i, c in enumerate(centroids_2d):
        axes[3].scatter(c[0], c[1], c=[colors[i]],
                        s=350, marker='*', zorder=5,
                        edgecolors='black', linewidth=0.8)
        axes[3].annotate(f'C{i+1}', (c[0], c[1]),
                         textcoords="offset points", xytext=(8, 6),
                         fontsize=11, fontweight='bold')
    axes[3].set_title('Step 4: Centroids\n= mean of each partition',
                      fontsize=11)
    axes[3].set_xlabel('PC1'); axes[3].set_ylabel('PC2')

    plt.tight_layout()
    plt.show()

# ─── Run for all datasets ────────────────────────────────
for name, (data, k) in datasets.items():
    visualize_espi_init(data, name, k)
```

```python
# ══════════════════════════════════════════════════════════
# PART 2 — PHASE B: DRIFT-BASED ITERATIONS (step by step)
# ══════════════════════════════════════════════════════════
# Each iteration:
#   1. Update centroids  →  compute drift (how far each moved)
#   2. Find active points  →  only points close to a drifted centroid
#   3. Reassign active points  →  check if they belong to a new cluster
# Repeat until no reassignments or centroids barely move.

def fit_with_history(data, k=3, max_iter=100, epsilon=1e-4):
    """Run KMeansCustom step-by-step, recording state at each iteration."""
    km = KMeansCustom(k=k, max_iterations=max_iter, epsilon=epsilon)
    km.centroids = km.espi_init(data)
    km.labels = km.full_assignment(data)

    history = [{
        'centroids': km.centroids.copy(),
        'labels':    km.labels.copy(),
        'active':    [],
        'drift':     np.zeros(k),
        'sse':       km.compute_sse(data),
        'reassigned': 0,
    }]

    for it in range(max_iter):
        old_centroids = km.centroids.copy()
        km.centroids = km.update_centroids(data)
        drift = km.compute_drift(old_centroids, km.centroids)
        active_indices = km.get_active_points(data, drift)

        reassigned = 0
        for idx in active_indices:
            dists = [km.compute_distance(data[idx], c) for c in km.centroids]
            new_label = np.argmin(dists)
            if new_label != km.labels[idx]:
                km.labels[idx] = new_label
                reassigned += 1

        history.append({
            'centroids':     km.centroids.copy(),
            'old_centroids': old_centroids,
            'labels':        km.labels.copy(),
            'active':        active_indices,
            'drift':         drift.copy(),
            'sse':           km.compute_sse(data),
            'reassigned':    reassigned,
        })

        if reassigned == 0 or np.max(drift) < epsilon:
            break

    return history


def plot_iterations(data, dataset_name, k=3):
    """Show selected iterations: clusters, active points, centroid arrows."""
    history = fit_with_history(data, k=k)
    total = len(history)

    # Pick iterations to display
    if total <= 6:
        show = list(range(total))
    else:
        show = sorted(set([0, 1, 2, total // 2, total - 2, total - 1]))

    n = len(show)
    cols = min(n, 3)
    rows = (n + cols - 1) // cols

    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)
    colors = plt.cm.Set1(np.linspace(0, 0.9, k))

    fig, axes = plt.subplots(rows, cols, figsize=(7*cols, 6*rows))
    axes = np.atleast_2d(np.array(axes).reshape(rows, cols))
    fig.suptitle(f'Phase B: Drift-Based Iterations — {dataset_name}\n'
                 f'(k={k}, converged in {total-1} iterations)',
                 fontsize=15, fontweight='bold', y=1.03)

    for idx, it in enumerate(show):
        r, c_ = divmod(idx, cols)
        ax = axes[r, c_]
        state = history[it]
        labels = state['labels']
        centroids_2d = pca.transform(state['centroids'])
        active = state['active']
        drift = state['drift']

        # Plot clusters
        for ci in range(k):
            mask = labels == ci
            ax.scatter(data_2d[mask, 0], data_2d[mask, 1],
                       c=[colors[ci]], alpha=0.35, s=20,
                       edgecolors='white', linewidth=0.2)

        # Highlight active points (black rings)
        if len(active) > 0:
            ax.scatter(data_2d[active, 0], data_2d[active, 1],
                       facecolors='none', edgecolors='black',
                       s=55, linewidth=0.9, alpha=0.6,
                       label=f'Active pts ({len(active)})')

        # Centroid drift arrows
        if 'old_centroids' in state:
            old_2d = pca.transform(state['old_centroids'])
            for ci in range(k):
                dx = centroids_2d[ci, 0] - old_2d[ci, 0]
                dy = centroids_2d[ci, 1] - old_2d[ci, 1]
                if np.sqrt(dx**2 + dy**2) > 0.01:
                    ax.annotate('', xy=centroids_2d[ci], xytext=old_2d[ci],
                                arrowprops=dict(arrowstyle='->', color='black',
                                                lw=1.5, ls='--'))

        # Centroid stars
        for ci in range(k):
            ax.scatter(centroids_2d[ci, 0], centroids_2d[ci, 1],
                       c=[colors[ci]], s=300, marker='*', zorder=5,
                       edgecolors='black', linewidth=0.8)

        # Title with stats
        if it == 0:
            title = f'After ESPI Init (Iter 0)\nSSE={state["sse"]:.2f}'
        else:
            title = (f'Iteration {it}\n'
                     f'SSE={state["sse"]:.2f}  |  '
                     f'Drift={drift.max():.4f}  |  '
                     f'Reassigned={state["reassigned"]}')
        ax.set_title(title, fontsize=9)
        ax.set_xlabel('PC1'); ax.set_ylabel('PC2')
        if len(active) > 0:
            ax.legend(fontsize=7, loc='upper right')

    # Hide empty subplots
    for idx in range(n, rows * cols):
        r, c_ = divmod(idx, cols)
        axes[r, c_].axis('off')

    plt.tight_layout()
    plt.show()

# ─── Run for all datasets ────────────────────────────────
for name, (data, k) in datasets.items():
    plot_iterations(data, name, k)
```

```python
# ══════════════════════════════════════════════════════════
# PART 3 — 3-ALGORITHM COMPARISON
# ══════════════════════════════════════════════════════════
# Compare:  Custom (ESPI+Drift)  vs  Alternate (Random+Furthest)  vs  Sklearn
# Across multiple k values on each dataset.

K_VALUES = [2, 3, 5, 10]

results = []

for ds_name, (data, _) in datasets.items():
    for k in K_VALUES:
        # ── Custom ───────────────────────────────────────
        t0 = time.time()
        custom = KMeansCustom(k=k)
        custom.fit(data)
        t_custom = time.time() - t0

        # ── Alternate ────────────────────────────────────
        np.random.seed(42)
        t0 = time.time()
        alt = KMeansAlternate(k=k)
        alt.fit(data)
        t_alt = time.time() - t0

        # ── Sklearn ──────────────────────────────────────
        t0 = time.time()
        sk = KMeans(n_clusters=k, random_state=42, n_init=1)
        sk.fit(data)
        t_sk = time.time() - t0

        results.append({
            'Dataset': ds_name, 'k': k,
            'Custom SSE': round(custom.sse_history[-1], 2),
            'Custom Iters': custom.n_iterations,
            'Custom Time (ms)': round(t_custom * 1000, 1),
            'Alternate SSE': round(alt.sse_history[-1], 2),
            'Alternate Iters': alt.n_iterations,
            'Alternate Time (ms)': round(t_alt * 1000, 1),
            'Sklearn SSE': round(sk.inertia_, 2),
            'Sklearn Iters': sk.n_iter_,
            'Sklearn Time (ms)': round(t_sk * 1000, 1),
        })

df_results = pd.DataFrame(results)
print("=" * 90)
print("FULL COMPARISON TABLE")
print("=" * 90)
print(df_results.to_string(index=False))
print()
```

```python
# ══════════════════════════════════════════════════════════
# PART 4 — SSE COMPARISON BAR CHARTS
# ══════════════════════════════════════════════════════════

fig, axes = plt.subplots(1, 3, figsize=(20, 5))
fig.suptitle('SSE Comparison: Custom vs Alternate vs Sklearn',
             fontsize=15, fontweight='bold', y=1.02)

for col, ds_name in enumerate(datasets.keys()):
    sub = df_results[df_results['Dataset'] == ds_name]
    x = np.arange(len(sub))
    w = 0.25

    axes[col].bar(x - w, sub['Custom SSE'],    w, label='Custom (ESPI+Drift)', color='#e74c3c')
    axes[col].bar(x,     sub['Alternate SSE'],  w, label='Alternate (Rand+Furthest)', color='#3498db')
    axes[col].bar(x + w, sub['Sklearn SSE'],    w, label='Sklearn', color='#2ecc71')

    axes[col].set_xticks(x)
    axes[col].set_xticklabels([f'k={k}' for k in sub['k']])
    axes[col].set_ylabel('SSE (lower is better)')
    axes[col].set_title(ds_name, fontsize=12)
    axes[col].legend(fontsize=7)

plt.tight_layout()
plt.show()
```

```python
# ══════════════════════════════════════════════════════════
# PART 5 — CONVERGENCE: SSE + ACTIVE POINTS OVER ITERATIONS
# ══════════════════════════════════════════════════════════

fig, axes = plt.subplots(2, 3, figsize=(20, 10))
fig.suptitle('Convergence Analysis — Custom KMeans (ESPI + Drift)',
             fontsize=15, fontweight='bold', y=1.02)

for col, (name, (data, k)) in enumerate(datasets.items()):
    history = fit_with_history(data, k=k)
    iters      = list(range(len(history)))
    sses       = [h['sse'] for h in history]
    n_active   = [len(h['active']) for h in history]
    reassigned = [h['reassigned'] for h in history]
    max_drift  = [h['drift'].max() for h in history]

    # ── Row 1: SSE curve ─────────────────────────────────
    ax1 = axes[0, col]
    ax1.plot(iters, sses, '-o', color='tab:red', markersize=4)
    ax1.set_xlabel('Iteration')
    ax1.set_ylabel('SSE')
    ax1.set_title(f'{name} — SSE Over Iterations', fontsize=12)
    ax1.grid(alpha=0.3)

    # ── Row 2: Active points + reassigned + drift ────────
    ax2 = axes[1, col]
    ax2.bar(iters, n_active, alpha=0.3, color='steelblue', label='Active Points')
    ax2.bar(iters, reassigned, alpha=0.6, color='tab:orange', label='Reassigned')
    ax2.set_xlabel('Iteration')
    ax2.set_ylabel('Count')

    ax3 = ax2.twinx()
    ax3.plot(iters, max_drift, '--^', color='tab:purple', markersize=4,
             alpha=0.8, label='Max Drift')
    ax3.set_ylabel('Max Drift', color='tab:purple')
    ax3.tick_params(axis='y', labelcolor='tab:purple')

    lines2, labels2 = ax2.get_legend_handles_labels()
    lines3, labels3 = ax3.get_legend_handles_labels()
    ax2.legend(lines2 + lines3, labels2 + labels3, fontsize=8, loc='upper right')
    ax2.set_title(f'{name} — Active Pts / Reassigned / Drift', fontsize=12)
    ax2.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ══════════════════════════════════════════════════════════
# PART 6 — FINAL CLUSTERS: SIDE-BY-SIDE (Custom vs Alternate vs Sklearn)
# ══════════════════════════════════════════════════════════

for ds_name, (data, k) in datasets.items():
    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)
    colors = plt.cm.Set1(np.linspace(0, 0.9, k))

    # Fit all three
    custom = KMeansCustom(k=k); custom.fit(data)
    np.random.seed(42)
    alt = KMeansAlternate(k=k); alt.fit(data)
    sk = KMeans(n_clusters=k, random_state=42, n_init=1); sk.fit(data)

    algos = [
        ('Custom\n(ESPI + Drift)', custom.labels, custom.centroids,
         custom.sse_history[-1], custom.n_iterations),
        ('Alternate\n(Random + Furthest)', alt.labels, alt.centroids,
         alt.sse_history[-1], alt.n_iterations),
        ('Sklearn\n(KMeans++)', sk.labels_, sk.cluster_centers_,
         sk.inertia_, sk.n_iter_),
    ]

    fig, axes = plt.subplots(1, 3, figsize=(20, 5.5))
    fig.suptitle(f'Final Clusters — {ds_name}  (k={k})',
                 fontsize=15, fontweight='bold', y=1.02)

    for col, (algo_name, labels, centroids, sse, n_it) in enumerate(algos):
        ax = axes[col]
        centroids_2d = pca.transform(centroids)

        for ci in range(k):
            mask = labels == ci
            ax.scatter(data_2d[mask, 0], data_2d[mask, 1],
                       c=[colors[ci]], alpha=0.5, s=25,
                       edgecolors='white', linewidth=0.2)
            ax.scatter(centroids_2d[ci, 0], centroids_2d[ci, 1],
                       c=[colors[ci]], s=300, marker='*', zorder=5,
                       edgecolors='black', linewidth=0.8)

        ax.set_title(f'{algo_name}\nSSE={sse:.2f}  |  Iters={n_it}', fontsize=11)
        ax.set_xlabel('PC1'); ax.set_ylabel('PC2')

    plt.tight_layout()
    plt.show()
```
