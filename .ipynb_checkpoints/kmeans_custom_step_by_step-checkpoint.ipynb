# K-Means Custom Algorithm Walkthrough

This notebook explains your `KMeansCustom` algorithm step by step and runs it on all three datasets in the project:

- Iris
- AI Index
- Earthquakes

It also repeats the walkthrough for several different values of `k` so you can see how the algorithm behaves as the number of clusters changes.


## What This Notebook Covers

The custom algorithm in `kmeans_custom.py` follows this flow:

1. **Percentile initialization**: create starting centroids between low and high feature percentiles.
2. **Initial full assignment**: assign every point to its closest centroid.
3. **Centroid update**: move each centroid to the mean of its assigned points.
4. **Drift computation**: measure how far each centroid moved.
5. **Outer-shell selection**: only recheck points near the cluster boundary.
6. **Partial reassignment**: reassign only those active points.
7. **Centroid refresh after shell reassignment**: align centroids to the new labels.
8. **Periodic full refresh**: every few iterations, reassign all points exactly as a safety check.
9. **Stopping rule**: stop when a refresh causes no changes and centroid drift is below `epsilon`.

The notebook records those stages so you can inspect:

- centroid positions
- number of active points
- number of reassigned points
- SSE after each iteration
- final cluster visualization in 2D PCA space


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display, Markdown
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

from kmeans_custom import KMeansCustom

plt.rcParams["figure.dpi"] = 120
plt.rcParams["axes.grid"] = True
````

````python
def load_datasets():
    iris = pd.read_csv("data/iris.csv")
    ai = pd.read_csv("data/ai_index.csv")
    eq = pd.read_csv("data/earthquakes.csv")
    return iris, ai, eq


def preprocess_iris(df):
    df = df.drop(columns=["species"]).dropna()
    return StandardScaler().fit_transform(df)


def preprocess_ai(df):
    df = df.select_dtypes(include=[np.number]).dropna()
    return StandardScaler().fit_transform(df)


def preprocess_eq(df):
    df = df[["magnitude", "depth", "latitude", "longitude"]].dropna()
    return StandardScaler().fit_transform(df)


iris_df, ai_df, eq_df = load_datasets()

datasets = {
    "Iris": preprocess_iris(iris_df),
    "AI Index": preprocess_ai(ai_df),
    "Earthquakes": preprocess_eq(eq_df),
}

dataset_summaries = pd.DataFrame(
    [
        {
            "Dataset": "Iris",
            "Rows": datasets["Iris"].shape[0],
            "Features": datasets["Iris"].shape[1],
            "Notes": "Classic flower dataset without the species label.",
        },
        {
            "Dataset": "AI Index",
            "Rows": datasets["AI Index"].shape[0],
            "Features": datasets["AI Index"].shape[1],
            "Notes": "Numeric country-level AI metrics only.",
        },
        {
            "Dataset": "Earthquakes",
            "Rows": datasets["Earthquakes"].shape[0],
            "Features": datasets["Earthquakes"].shape[1],
            "Notes": "Magnitude, depth, latitude, and longitude.",
        },
    ]
)

dataset_summaries
````

````python
class KMeansCustomTracer(KMeansCustom):
    """
    Extends KMeansCustom with a trace log so the notebook can show
    what the algorithm does at each stage.
    """

    def _record_step(
        self,
        trace,
        iteration,
        step_name,
        description,
        data,
        drift=None,
        active_indices=None,
        reassigned=None,
        did_full_refresh=False,
    ):
        labels = None if self.labels is None else self.labels.copy()
        centroids = None if self.centroids is None else self.centroids.copy()
        cluster_sizes = None

        if labels is not None:
            cluster_sizes = [int(np.sum(labels == i)) for i in range(self.k)]

        trace.append(
            {
                "iteration": iteration,
                "step": step_name,
                "description": description,
                "centroids": centroids,
                "labels": labels,
                "cluster_sizes": cluster_sizes,
                "max_drift": None if drift is None else float(np.max(drift)),
                "active_points": 0 if active_indices is None else int(len(active_indices)),
                "reassigned": None if reassigned is None else int(reassigned),
                "did_full_refresh": bool(did_full_refresh),
                "sse": None if labels is None or centroids is None else float(self.compute_sse(data)),
            }
        )

    def trace_fit(self, data):
        data = np.asarray(data, dtype=float)
        trace = []

        self.centroids = self.percentile_init(data)
        self.labels = None
        self._record_step(
            trace,
            iteration=0,
            step_name="percentile_init",
            description="Initial centroids are placed between low and high feature percentiles.",
            data=data,
        )

        self.labels = self.full_assignment(data)
        self._record_step(
            trace,
            iteration=0,
            step_name="initial_full_assignment",
            description="Every point is assigned to the nearest starting centroid.",
            data=data,
        )

        self.centroids = self.update_centroids(data)
        self._record_step(
            trace,
            iteration=0,
            step_name="initial_centroid_update",
            description="Centroids move to the mean of their assigned cluster members.",
            data=data,
        )

        self.sse_history = []
        self.reassigned_history = []
        self.n_iterations = 0

        for iteration in range(1, self.max_iterations + 1):
            old_centroids = self.centroids.copy()

            self.centroids = self.update_centroids(data)
            drift = self.compute_drift(old_centroids, self.centroids)
            self._record_step(
                trace,
                iteration=iteration,
                step_name="centroid_update_and_drift",
                description="Current labels are used to refresh centroids and compute centroid drift.",
                data=data,
                drift=drift,
            )

            active_indices = self.get_active_points(data, drift)
            self._record_step(
                trace,
                iteration=iteration,
                step_name="outer_shell_selection",
                description="Only boundary points are selected for possible reassignment.",
                data=data,
                drift=drift,
                active_indices=active_indices,
            )

            reassigned = 0
            for idx in active_indices:
                point = data[idx]
                distances = [self.compute_distance(point, c) for c in self.centroids]
                new_label = int(np.argmin(distances))

                if new_label != self.labels[idx]:
                    self.labels[idx] = new_label
                    reassigned += 1

            self._record_step(
                trace,
                iteration=iteration,
                step_name="partial_reassignment",
                description="Only the active outer-shell points are checked against all centroids.",
                data=data,
                drift=drift,
                active_indices=active_indices,
                reassigned=reassigned,
            )

            shell_old_centroids = self.centroids.copy()
            self.centroids = self.update_centroids(data)
            drift = self.compute_drift(shell_old_centroids, self.centroids)
            self._record_step(
                trace,
                iteration=iteration,
                step_name="post_shell_centroid_refresh",
                description="Centroids are updated again so they match the new labels after shell reassignment.",
                data=data,
                drift=drift,
                active_indices=active_indices,
                reassigned=reassigned,
            )

            did_full_refresh = (
                self.full_refresh_every > 0
                and iteration % self.full_refresh_every == 0
            )

            if did_full_refresh:
                refreshed_labels = self.full_assignment(data)
                extra_changes = int(np.sum(refreshed_labels != self.labels))
                self.labels = refreshed_labels
                reassigned += extra_changes

                refresh_old_centroids = self.centroids.copy()
                self.centroids = self.update_centroids(data)
                drift = self.compute_drift(refresh_old_centroids, self.centroids)

                self._record_step(
                    trace,
                    iteration=iteration,
                    step_name="periodic_full_refresh",
                    description="A full exact reassignment is done for safety, then centroids are updated again.",
                    data=data,
                    drift=drift,
                    reassigned=reassigned,
                    did_full_refresh=True,
                )

            self.sse_history.append(self.compute_sse(data))
            self.reassigned_history.append(reassigned)
            self.n_iterations += 1

            self._record_step(
                trace,
                iteration=iteration,
                step_name="metrics_recorded",
                description="SSE and reassignment counts are stored after labels and centroids are aligned.",
                data=data,
                drift=drift,
                reassigned=reassigned,
                did_full_refresh=did_full_refresh,
            )

            if did_full_refresh and reassigned == 0 and np.max(drift) < self.epsilon:
                self._record_step(
                    trace,
                    iteration=iteration,
                    step_name="stop",
                    description="The algorithm stops because the refresh changed nothing and drift is below epsilon.",
                    data=data,
                    drift=drift,
                    reassigned=reassigned,
                    did_full_refresh=True,
                )
                break

        return trace
````

````python
def trace_to_frame(trace):
    return pd.DataFrame(
        [
            {
                "iteration": row["iteration"],
                "step": row["step"],
                "description": row["description"],
                "active_points": row["active_points"],
                "reassigned": row["reassigned"],
                "max_drift": row["max_drift"],
                "sse": row["sse"],
                "cluster_sizes": row["cluster_sizes"],
                "full_refresh": row["did_full_refresh"],
            }
            for row in trace
        ]
    )


def summarize_run(model, data, dataset_name, k):
    final_sse = float(model.compute_sse(data))
    return pd.DataFrame(
        [
            {
                "Dataset": dataset_name,
                "k": k,
                "Iterations": model.n_iterations,
                "Final SSE": round(final_sse, 4),
                "Final cluster sizes": [int(np.sum(model.labels == i)) for i in range(model.k)],
            }
        ]
    )


def plot_run(model, data, dataset_name, k):
    pca = PCA(n_components=2)
    data_2d = pca.fit_transform(data)

    fig, axes = plt.subplots(1, 3, figsize=(16, 4))
    fig.suptitle(f"{dataset_name} | k={k}", fontsize=14, fontweight="bold")

    scatter = axes[0].scatter(
        data_2d[:, 0],
        data_2d[:, 1],
        c=model.labels,
        cmap="tab10",
        s=28,
        alpha=0.85,
    )
    axes[0].set_title("Final Clusters (PCA)")
    axes[0].set_xlabel("PC1")
    axes[0].set_ylabel("PC2")
    fig.colorbar(scatter, ax=axes[0], fraction=0.046, pad=0.04)

    axes[1].plot(
        range(1, len(model.sse_history) + 1),
        model.sse_history,
        marker="o",
        color="#1f77b4",
    )
    axes[1].set_title("SSE by Iteration")
    axes[1].set_xlabel("Iteration")
    axes[1].set_ylabel("SSE")

    axes[2].plot(
        range(1, len(model.reassigned_history) + 1),
        model.reassigned_history,
        marker="o",
        color="#d62728",
    )
    axes[2].set_title("Points Reassigned")
    axes[2].set_xlabel("Iteration")
    axes[2].set_ylabel("Count")

    plt.tight_layout()
    plt.show()


def run_walkthrough(dataset_name, data, k, trace_rows=14):
    display(Markdown(f"## {dataset_name} with k = {k}"))

    model = KMeansCustomTracer(k=k)
    trace = model.trace_fit(data)
    trace_df = trace_to_frame(trace)

    display(summarize_run(model, data, dataset_name, k))
    display(
        trace_df[
            [
                "iteration",
                "step",
                "description",
                "active_points",
                "reassigned",
                "max_drift",
                "sse",
                "cluster_sizes",
                "full_refresh",
            ]
        ].head(trace_rows)
    )

    if len(trace_df) > trace_rows:
        display(
            Markdown(
                f"Showing the first {trace_rows} trace rows out of {len(trace_df)} total rows for this run."
            )
        )

    plot_run(model, data, dataset_name, k)
    return model, trace_df
````

## Experiment Settings

The cell below uses four different `k` values:

- `2`
- `5`
- `10`
- `20`

That gives you 12 runs in total across the three datasets.

If the full output is too long, reduce `K_VALUES` or call `run_walkthrough(...)` for one dataset at a time.


````python
K_VALUES = [2, 5, 10, 20]
````

````python
all_run_results = {}
summary_rows = []

for dataset_name, data in datasets.items():
    for k in K_VALUES:
        model, trace_df = run_walkthrough(dataset_name, data, k)
        all_run_results[(dataset_name, k)] = {
            "model": model,
            "trace": trace_df,
        }
        summary_rows.append(
            {
                "Dataset": dataset_name,
                "k": k,
                "Iterations": model.n_iterations,
                "Final SSE": round(float(model.compute_sse(data)), 4),
                "Total reassignments": int(sum(model.reassigned_history)),
            }
        )
````

````python
summary_df = pd.DataFrame(summary_rows).sort_values(["Dataset", "k"]).reset_index(drop=True)
summary_df
````
